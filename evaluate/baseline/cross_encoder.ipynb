```python
from typing import List
from tester import MetamorphicTesting
from embeddings import HuggingfaceEmbeddings
import jsonlines

cross_encdoer_models = ['cross-encoder/nli-deberta-v3-base', 'cross-encoder/quora-distilroberta-base', 'FacebookAI/roberta-large-mnli']

metamorphic = ['word_swap', 'obj_sub', 'verb_sub', 'nega_exp', 'word_del', 'num_sub', 'err_translate', 'err_nli']
distance_metrics = ['cosine', 'euclidean', 'person', 'manhattan', 'lancewilliams', 'mahalanobis', 'braycurtis']


def load_dataset(path) -> List[List[str]]:
    data = []
    with jsonlines.open(path) as f:
        for line in f:
            data.append([line['sentence1'], line['sentence2'], line['sentence3']])
    return data


if __name__ == "__main__":
    for model_name in cross_encdoer_models:
      embedding = HuggingfaceEmbeddings(model_name=model_name)
      vector_db = 'Annoy'
      distance_metric = None
      mt = MetamorphicTesting(embedding, distance_metric, vector_db)
      for me in metamorphic:
          dataset = load_dataset('data/MeTMaP/dataset/normal/'+me+'.jsonl')
          results = mt.run_predictions(dataset, subset=me, reverse=True)
          # evaluates = mt.evaluate_baselines(results)


```
